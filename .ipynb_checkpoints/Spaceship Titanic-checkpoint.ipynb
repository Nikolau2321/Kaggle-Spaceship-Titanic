```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

# Prepare the training data

```python
df_train = pd.read_csv(r'G:\Coding ninjas\Kaggle projects & submissions\Spaceship Titanic\train.csv')
df_train
```

```python
df_train.isnull().sum()
```

```python
df_train_hp = pd.get_dummies(df_train['HomePlanet'], drop_first = True)
df_train.drop('HomePlanet', axis = 1, inplace = True)
df_train2 = pd.concat([df_train, df_train_hp], axis = 1)
df_train2.head()
```

```python
df_train2['CryoSleep'].value_counts()
```

```python
df_train2['CryoSleep'].mode().values[0]
```

```python
df_train2['CryoSleep'].replace((np.nan, df_train2['CryoSleep'].mode().values[0]), inplace = True)
df_train2['CryoSleep'] = df_train2['CryoSleep'].astype(str)
df_train2['CryoSleep'].replace({'False':0, 'True':1}, inplace = True)

df_train2['CryoSleep'].isnull().sum()
```

```python
df_train2['VIP'].replace((np.nan, df_train2['VIP'].mode().values[0]), inplace = True)
df_train2['VIP'] = df_train2['VIP'].astype(str)
df_train2['VIP'].replace({'False':0, 'True':1}, inplace = True)

df_train2['VIP'].isnull().sum()
```

```python
df_train2['Transported'] = df_train2['Transported'].astype(str)
df_train2['Transported'].replace({'False':0, 'True':1}, inplace = True)
```

```python
df_train_d = pd.get_dummies(df_train2['Destination'], drop_first = True)
df_train2.drop('Destination', axis =1, inplace = True)
df_train3 = pd.concat([df_train2, df_train_d], axis = 1)
df_train3.head()
```

```python
df_train3['Cabin'].isnull().sum()
```

```python
df_train3['Cabin'].mode().values[0]
```

```python
df_train3['Cabin'].replace((np.nan, df_train3['Cabin'].mode().values[0]), inplace = True)
```

```python
df_train3['Cabin'] = df_train3['Cabin'].apply(lambda x: x.split('/')[2])
df_train3.head()
```

```python
df_train3['Cabin'].replace({'P':0, 'S':1}, inplace = True)
```

```python
df_train3.drop(['Name', 'PassengerId'], axis = 1, inplace = True)
df_train3.head()
```

```python
df_train3.isnull().sum()
```

```python
df_train3['Age'].replace((np.nan, df_train3['Age'].median()), inplace = True)
df_train3['RoomService'].replace((np.nan, df_train3['RoomService'].mean()), inplace = True)
df_train3['FoodCourt'].replace((np.nan, df_train3['FoodCourt'].mean()), inplace = True)
df_train3['ShoppingMall'].replace((np.nan, df_train3['ShoppingMall'].mean()), inplace = True)
df_train3['Spa'].replace((np.nan, df_train3['Spa'].mean()), inplace = True)
df_train3['VRDeck'].replace((np.nan, df_train3['VRDeck'].mean()), inplace = True)

df_train3.isnull().sum()
```

```python
Xtrain = df_train3.drop('Transported', axis = 1)
Xtrain.shape
```

```python
Ytrain = df_train3['Transported']
```

# Prepare the testing data 

```python
df_test = pd.read_csv(r'G:\Coding ninjas\Kaggle projects & submissions\Spaceship Titanic\test.csv')
df_test
```

```python
pass_id = df_test['PassengerId']
```

```python
df_test_hp = pd.get_dummies(df_test['HomePlanet'], drop_first = True)

df_test.drop('HomePlanet', axis = 1, inplace = True)
df_test2 = pd.concat([df_test, df_test_hp], axis = 1)
df_test2.head()
```

```python


df_test2['CryoSleep'].replace((np.nan, df_test2['CryoSleep'].mode().values[0]), inplace = True)
df_test2['CryoSleep'] = df_test2['CryoSleep'].astype(str)
df_test2['CryoSleep'].replace({'False':0, 'True':1}, inplace = True)


df_test2['VIP'].replace((np.nan, df_test2['VIP'].mode().values[0]), inplace = True)
df_test2['VIP'] = df_test2['VIP'].astype(str)
df_test2['VIP'].replace({'False':0, 'True':1}, inplace = True)

df_test2.isnull().sum()
```

```python
df_test_d = pd.get_dummies(df_test2['Destination'], drop_first = True)

df_test2.drop('Destination', axis =1, inplace = True)
df_test3 = pd.concat([df_test2, df_test_d], axis = 1)

df_test3.head()
```

```python
df_test3['Cabin'].replace((np.nan, df_test3['Cabin'].mode().values[0]), inplace = True)
df_test3['Cabin'] = df_test3['Cabin'].apply(lambda x: x.split('/')[2])
df_test3['Cabin'].replace({'P':0, 'S':1}, inplace = True)
df_test3['Cabin'].isnull().sum()
```

```python
df_test3.drop(['Name', 'PassengerId'], axis = 1, inplace = True)
```

```python
df_test3['Age'].replace((np.nan, df_test3['Age'].median()), inplace = True)
df_test3['RoomService'].replace((np.nan, df_test3['RoomService'].mean()), inplace = True)
df_test3['FoodCourt'].replace((np.nan, df_test3['FoodCourt'].mean()), inplace = True)
df_test3['ShoppingMall'].replace((np.nan, df_test3['ShoppingMall'].mean()), inplace = True)
df_test3['Spa'].replace((np.nan, df_test3['Spa'].mean()), inplace = True)
df_test3['VRDeck'].replace((np.nan, df_test3['VRDeck'].mean()), inplace = True)
df_test3.isnull().sum()
```

```python
Xtest = df_test3
Xtest.shape
```

# Training the model

```python
#by using Logistic Regression
logReg = LogisticRegression(penalty = 'l2', max_iter = 100000)
grid1 = {'solver':['lbfgs', 'newton-cg', 'sag']}

bestLogReg = GridSearchCV(logReg, grid1, cv= 5)
bestLogReg.fit(Xtrain, Ytrain)
bestLogReg.best_estimator_
```

```python
bestLogReg.score(Xtrain, Ytrain)
```

```python
#by using Random Forests
ranFor = RandomForestClassifier(n_estimators = 100, criterion = 'gini')
grid2 = {'max_depth':[5,6,7]}

bestRanFor = GridSearchCV(ranFor, grid2, cv = 5)
bestRanFor.fit(Xtrain, Ytrain)
bestRanFor.best_estimator_
```

```python
bestRanFor.score(Xtrain, Ytrain)
```

```python
Ypred = bestRanFor.predict(Xtest)
Ypred
```

```python

Ypred_labelled = Ypred.astype(bool)


Ypred_labelled 
```

```python
Ypred_labelled.dtype
```

```python
df_preds = pd.DataFrame({'PassengerId': pass_id, 'Transported': Ypred_labelled})
df_preds
```

```python
df_preds.to_csv('Spaceship_Titanic_pred3.csv', index = False)
```
